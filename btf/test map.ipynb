```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import frozen_lake
```

```python
key = jax.random.PRNGKey(0)
p = 0.5
for scale in range(1, 6):
    print(f"generating map of size {2**scale}x{2**scale}")
    frozen_lake.generate_chunks(key, scale, p)
    %timeit frozen_lake.generate_chunks(key, scale, p)
    print("Example maps:")
    
    plt.figure(figsize=(10, 2))
    for j, key in enumerate(jax.random.split(key, 4)):
        plt.subplot(1, 4, j+1)
        plt.imshow(frozen_lake.generate_chunks(key, scale, p)[0], cmap='gray')
    plt.show()
```
